```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
```

```python
response= requests.get('http://10.143.182.228:9898/jiodata/stats?value=5')
df=pd.DataFrame(response.json())
df
```

```python
x = df['PG_NAME']
y = df['SUCCESS']

plt.figure(figsize=(25, 6))  
plt.bar(x, y)
plt.xlabel('Payment Gateways',fontsize=18)
plt.ylabel('Successful Recharges',fontsize=18)
plt.title('Successful Recharges by Payment Gateways',fontsize=25)
plt.show()
```

```python

x = df['PG_NAME']
y = df['FAILURES']

plt.figure(figsize=(25, 6))  
plt.bar(x, y)
plt.xlabel('Payment Gateways',fontsize=18)
plt.ylabel('Failed Recharges',fontsize=18)
plt.title('Failed Recharges by Payment Gateways',fontsize=25)
plt.show()
```

```python
x = df['PG_NAME']
y = df['TOTAL RECHARGES']

plt.figure(figsize=(25, 6))  
plt.bar(x, y)
plt.xlabel('Payment Gateways',fontsize=18)
plt.ylabel('Not confirmed Recharges',fontsize=18)
plt.title('Total Recharges by Payment Gateways',fontsize=25)
plt.show()
```

```python
x = df['PG_NAME']
y = df['TOTAL RECHARGES']

plt.figure(figsize=(25, 6))  
plt.bar(x, y)
plt.xlabel('Payment Gateways',fontsize=18)
plt.ylabel('Total Recharges',fontsize=18)
plt.title('Total Recharges by Payment Gateways',fontsize=25)
plt.show()
```

```python
x = df['PG_NAME']
width = 0.2
fig, ax = plt.subplots(figsize=(25, 6))

# Generate the bar positions with gaps between groups
n_groups = len(df)
bar_positions = [i * 1.2 for i in range(n_groups)]

# Plot each bar group
ax.bar([pos - 1.5 * width for pos in bar_positions], df['SUCCESS'], width=width, align='center', label='Success')
ax.bar([pos - 0.5 * width for pos in bar_positions], df['FAILURES'], width=width, align='center', label='Failure')
ax.bar([pos + 0.5 * width for pos in bar_positions], df['WIP-NOTCONFIRMED'], width=width, align='center', label='Not Confirmed')
ax.bar([pos + 1.5 * width for pos in bar_positions], df['TOTAL RECHARGES'], width=width, align='center', label='Total')

# Set x-ticks and labels
ax.set_xticks(bar_positions)
ax.set_xticklabels(x)

ax.set_xlabel('Payment Gateways', fontsize=25)
ax.set_ylabel('Number of Recharges', fontsize=25)
ax.set_title('Success, Failure, Unconfirmed, and Total Recharges by Payment Gateway', fontsize=35)

ax.legend()

plt.tight_layout()
plt.show()
```
